```python
import pandas as pd
```

```python
# Load your CSV file
df = pd.read_csv(r"audio_to_text_similarity_refined.csv")  # Replace with your actual file path

# Define the emotion columns
emotion_cols = ['Stimulating', 'Playful', 'Soothing', 'Sensory-Calming', 
                'Grounding', 'Focusing', 'Transitional', 'Anxiety-Reduction']
```

```python
# Function to get top 3 emotions
def get_top_emotions(row):
    top_emotions = row[emotion_cols].sort_values(ascending=False).index.tolist()
    return pd.Series(top_emotions[:3], index=['emotion1', 'emotion2', 'emotion3'])

# Apply the function to each row and join the result to the original dataframe
df[['emotion1', 'emotion2', 'emotion3']] = df.apply(get_top_emotions, axis=1)
```
